Analise de budget variance, forecast e run rate da camada Gold.

```python
from pathlib import Path
import pandas as pd
sources = list(Path('../data/gold/indicadores_fpa').glob('fpa_*.csv'))
fpa = pd.concat([pd.read_csv(path) for path in sources], ignore_index=True)
fpa[['loja_id', 'receita_realizada', 'receita_orcada', 'forecast_mes', 'run_rate_anual', 'status_semaforo']].sort_values('receita_realizada', ascending=False)
```
