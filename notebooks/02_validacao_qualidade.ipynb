# Qualidade de Dados
Inspecao dos relatorios produzidos pela validacao Bronze.

```python
from pathlib import Path
import json
import pandas as pd
reports = []
for path in Path('../data/bronze/vendas').glob('ano=*/mes=*/dia=*/relatorio_qualidade_*.json'):
    reports.append(json.loads(path.read_text(encoding='utf-8')))
pd.DataFrame(reports)[['data_referencia', 'total_registros', 'registros_validos', 'registros_invalidos', 'taxa_aprovacao']].sort_values('data_referencia')
```
