# Analise Exploratoria Raw
Le arquivos reais emitidos pelo PDV e resume volume por loja e turno.

```python
from pathlib import Path
import json
import pandas as pd
raw_files = list(Path('../data/raw/vendas').glob('loja=*/ano=*/mes=*/dia=*/*.json'))
rows = []
for file in raw_files:
    payload = json.loads(file.read_text(encoding='utf-8'))
    rows.append({'loja_id': payload['loja_id'], 'turno': payload['turno'], 'registros': len(payload['vendas'])})
pd.DataFrame(rows).groupby(['loja_id', 'turno'], as_index=False)['registros'].sum()
```
